# Explore Data Distribution

## Sale Price

```python
df['SalePrice'].plot(kind = 'hist', bins = 30, alpha = 0.7, color = 'blue', )

plt.title("Histogram of SalePrice")
plt.xlabel("SalePrice")
plt.ylabel("Frequency")

plt.show()
```

### Descriptive Statistics

```python
mean = df['SalePrice'].mean()
median = df['SalePrice'].median()
std = df['SalePrice'].std()

print(f"\nmean of sale price is: {mean}")
print(f"median of sale price is: {median}")
print(f"standard deviation of sale price is: {std}")
```

```python
f"""
# Sale price distribution
The Histogram of SalePrice shows the distribution of sale prices is right-skewed or rather positively-skewed. The majority of the prices are on the lower side of the histogram compared to few prices forming the tail on high sale prices.
The mean of sale price is: {mean} with a standard deviation of {std}. The standard deviation is so little and indicates minimal deviation from the mean sale price. The mean and the median are slightly different with the mean being the highest.
Also, the median value indicates that half of the houses sold less than {median} with the other half more than the {median}
"""
```

# Explore Differences between Subsets 

```python
# Replace None with appropriate code
below_average_condition = df[df['OverallCond'] < 5]
average_condition = df[df['OverallCond'] == 5]
above_average_condition = df[df['OverallCond'] > 5]
```

```python
# Run this cell without changes

# Check that all of them still have 80 columns
assert below_average_condition.shape[1] == 80
assert average_condition.shape[1] == 80
assert above_average_condition.shape[1] == 80

# Check the numbers of rows of each subset
assert below_average_condition.shape[0] == 88
assert average_condition.shape[0] == 821
assert above_average_condition.shape[0] == 551
```

```python
# Run this cell without changes

# Set up plot
fig, ax = plt.subplots(figsize=(15,5))

# Create custom bins so all are on the same scale
bins = range(df["SalePrice"].min(), df["SalePrice"].max(), int(df["SalePrice"].median()) // 20)

# Plot three histograms, with reduced opacity (alpha) so we
# can see them overlapping
ax.hist(
    x=above_average_condition["SalePrice"],
    label="above average condition",
    bins=bins,
    color="cyan",
    alpha=0.5
)
ax.hist(
    x=average_condition["SalePrice"],
    label="average condition",
    bins=bins,
    color="gray",
    alpha=0.3
)
ax.hist(
    x=below_average_condition["SalePrice"],
    label="below average condition",
    bins=bins,
    color="yellow",
    alpha=0.5
)

# Customize labels
ax.set_title("Distributions of Sale Price Grouped by Condition")
ax.set_xlabel("Sale Price")
ax.set_ylabel("Number of Houses")
ax.legend();
```

```python
# Replace None with appropriate text
"""
Properties in above average condition command higher sale prices compared to those in average or below average condition.
The distribution for above average condition is shifted more towards the right with a higher concentration of sale prices in the higher price ranges.
The distribution for below average condition is skewed towards lower sale prices with a significant portion of properties selling for prices below $200,000.
The distributions for above average and average condition are more dispersed and have longer tails towards higher sale prices compared to the below average condition distribution.
"""
```
